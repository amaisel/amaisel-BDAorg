### Problem 1.1 (1/2 point):

> *How many possible binary logical operations are there?*  Don't just report a number; briefly explain your reasoning.

There are $2^{2^n}$ possible binary logical operations, where n is the number of variables being introduced. This is intuitive, since any single vairable (i.e. A) can take on one of two values, $A$ and $\overline A$, and a variable B also has these two states, B and $\overline B$. When we combine these, we have $2^{4} = 16$ possible situations, (i.e. $A$, $A \land B$, $A \land \overline B$, etc.) If a third variable C were introduced, we would have $2^8 =256$ options, and so on. Thus, we have $2^{2^n}$ combinations for a given number of variables to be evaluated.
